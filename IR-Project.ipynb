# Movie-recommendation-System

```python
import numpy as np
import pandas as pd

movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')

# merging both datasets on basis of title
movies=movies.merge(credits,on='title')


#genre
#id
#keywords
#title
#overview
#release_date
#cast

movies=movies[['movie_id','title','release_date','overview','genres','keywords','cast','crew']]

#missing data check
movies.isnull().sum()
#in case any missing -  
movies.dropna(inplace=True)

#check duyplicate rows
movies.duplicated().sum()
```

```python
import ast

#converting genere dictionary to array of genre name
def congenre(obj):
    genre=[]
    for i in ast.literal_eval(obj):
        genre.append(i['name'])
    return genre


 # to convert genre dictionary string as integer list
# gen=ast.literal_eval(movies.iloc[0].genres)

# congenre(gen)
```

```python
movies['genres']=movies['genres'].apply(congenre) #already done that's why bad output

movies['keywords']=movies['keywords'].apply(congenre)
```

```python
#appending 3 cast names only
def concast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L


def Director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python

movies['crew']=movies['crew'].apply(Director)
movies['cast']=movies['cast'].apply(concast)

movies['cast']

```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())

movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])


#we need to remove spaces from all the column data

movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])#we need to remove spaces from all the column data
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']

movies.head()
```

```python
new_df=movies[['movie_id','title','tag']]
new_df['title']=new_df['title'].apply(lambda x:x.lower())
#as we see lists in tag col so let's make it string
new_df['tag']=new_df['tag'].apply(lambda x: " ".join(x))
```

```python
new_df['tag']=new_df['tag'].apply(lambda x:x.lower())
#converting all upper letters to lower

new_df
```

```python
# we need to make tag text into vectors of data and then we will show nearest 10 vectors

# we will use bag of words technique and k nearestneighbour

# we will merge all tags and then will make 5000 word dataset(most frequent words) (choose less words to perform better)
# then we will check no of words in movie to be recommended  
#  we will get shape(no of movies,no of freq word dataset)



### also stopwords are avoided (helping verbs,. and all that)

# we can use scikit-learn for above purpose
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english')
#max_features is no of words to take from bag

vector_movies=cv.fit_transform(new_df['tag']).toarray()

vector_movies.shape  # no of movies,no of words we chose

# cv.get_feature_names()    ##to get names
```

```python
# we can see lots of words are similiar so we need to apply stamming (make similiar words as any one )
# we need nltk 


# !pip install nltk

from nltk.stem.porter import  PorterStemmer

ps=PorterStemmer()

def stem(text):
    y=[]
    for  i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)




new_df['tag']=new_df['tag'].apply(stem)
```

```python
#now we need to recreate words



cv=CountVectorizer(max_features=5000,stop_words='english')
#max_features is no of words to take from bag
vectors=cv.fit_transform(new_df['tag']).toarray()
```

```python
# cv.get_feature_names()
# !pip install sklearn
```

```python
#  now to give nearest 10 ,we need to find distance in between and for vector we need cosine distance(angle between two vectors)
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(vectors)
```

```python
# sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])  # enum so that we don't loose index if used to sort
# sorted(similarity[0],reverse=True) #matrix with diagonal 1


result=pd.DataFrame()

result['movie_id']=movies['movie_id']
result['title']=movies.title+"("+movies['release_date']+")"
result['genre']=movies['genres'].apply(lambda x:",".join(x))
result['cast']=movies['cast'].apply(lambda x:",".join(x))
result['overview']=movies['overview'].apply(lambda x:" ".join(x))
```

```python
def recommend(mov):
    
        n=10 #no of movies user desire to be recommended
        movies_index=new_df[new_df['title']==mov.lower()].index[0]
        distance=similarity[movies_index]
        movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:n+1]
        L=[]
        for i in movies_list:
            t=result.iloc[i[0]]
            df=pd.DataFrame()
            df=[t.movie_id,t.title,t.genre,t.cast,t.overview]
            L.append(df)
        
        header=['Movie_id','Title(Release Date)','Genres','Top 3 Cast','A Brief Overview']
        Movie=pd.DataFrame(L,columns=header)
        return Movie
```
